```python
import cv2 
import numpy as np 
from google.colab.patches import cv2_imshow
```

```python
image1 = cv2.imread('kumbang.jpg')
```

```python
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
```

```python
ret, thresh1 = cv2.threshold(img, 50,250, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(img, 50, 250, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(img, 50, 250, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(img, 50, 250, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(img, 50, 250, cv2.THRESH_TOZERO_INV)
```
